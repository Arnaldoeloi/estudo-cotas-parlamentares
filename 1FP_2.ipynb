# 2.0 Data Processing

Primeiro, precisaremos concatenar os dados de 2019 e 2020 para que possamos, mais tarde visualizar a comparação dos gastos.

```
df_final = pd.concat([df_2019,df_2020], axis=0)
```

Vejamos como nossa base de dados ficou:

```
df_final.tail()
```

Podemos executar um **.info()** em nosso dataset para verificar o formato e colunas de nossos dados.

```
df_final.info()
```

Utilizaremos, de todas, as seguintes:

```
df = df_final.iloc[:,[0,6,9,11,19,21]].copy()
df
```

Agora, renomearemos as colunas para facilitar o uso.

```
df.columns = ['nome', 'partido', 'tipo', 'desc', 'valor', 'ano']
df
```

Como faremos operações com o valor, é válido converter a coluna para numérico.

```
df.valor = pd.to_numeric(df.valor, errors="coerce")
df.info()
```

Vamos agora definir como index o partido e ano, com as colunas **Tipo** com todos os valores somados para cada integrante do partido.

```
df_reshaped = pd.pivot_table(df, index=['partido','ano'], columns='tipo', values='valor', aggfunc=sum,fill_value=0).copy()
```

```
df_reshaped.head(10)
```

Utilizaremos uma nova coluna com a soma dos valores anteriores mostrando o total.

```
df_reshaped['total'] = df_reshaped.sum(axis=1)
```

Como nosso trabalho de pesquisa busca saber sobre quantidade de salários mínimos, vamos definir uma função que retorne este valor:

```
 def salarios(x):
   return x/1045
```

Agora, criamos uma nova coluna com a quantidade de salário utilizando o **map** na coluna **total**.

```
df_reshaped['salarios'] = df_reshaped.total.map(salarios)
```

```
df_reshaped
```

Agora, finalmente faremos o gráfico para visualizar os dados capturados. Uma ótima opção é o gráfico de barras, onde cada barra é um ano dividido para cada partido.

```
ax = df_reshaped.unstack().plot(kind="bar",fontsize=20, y="salarios", figsize=(40,12), rot=45, color=["#2ba152", "#f28322"], title="Quantos salários mínimos foram necessários para pagar as cotas dos deputados? (2019 vs 2020)", xlabel="Partido", ylabel="Salários")
ax.title.set_size(24)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
ax.yaxis.label.set_size(20)
ax.xaxis.label.set_size(20)

ax
```
